```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\lenovo\Documents\ML Practice\homeprices.csv")
df
```

```python
# Plot scatter plot - to get an idea on distribution of data points - we use %matplotlib inline magic in order to draw plots
```

```python
%matplotlib inline
plt.scatter(df.area,df.price)
```

```python
# we give label to the x and y axis and can perform different operations in scatter
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)')
plt.ylabel('price (US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
# will create an object or linear regression 
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
# .fit() fitting the data means you are training the linear regression model using available data points
```

```python
reg.predict([[3300]])
```

```python
# and we came up with this as for linear eq we have y=mx+b it will calculate m and b i.e coefficient & intercept and 
# price = m*area+b
```

```python
reg.coef_
```

```python
reg.intercept_
```
